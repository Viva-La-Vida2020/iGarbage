```python
from PIL import Image
class Resize(object):
    def __init__(self, size, interpolation=Image.BILINEAR):
        self.size = size
        self.interpolation = interpolation

    def __call__(self, img):
        # 计算纵横比
        ratio = self.size[0] / self.size[1]
        w, h = img.size

        if w / h < ratio:
            # 如果图像的宽高比小于目标纵横比，执行以下操作
            t = int(h * ratio)
            w_padding = (t - w) // 2
            img = img.crop((-w_padding, 0, w+w_padding, h))
        else:
            # 如果图像的宽高比大于目标纵横比，执行以下操作
            t = int(w / ratio)
            h_padding = (t - h) // 2
            img = img.crop((0, -h_padding, w, h+h_padding))

        # 调整图像大小
        img = img.resize(self.size, self.interpolation)

        return img


# 示例调用
resize = Resize((256, 256))  # 创建Resize对象，设置目标大小
image = Image.open('C:\\Users\\Administrator\\Desktop\\垃圾分类\\data\\garbage_classify\\train_data\img_1.jpg' )  # 打开图像文件
resized_image = resize(image)  # 调整图像大小
resized_image.show()
```

```python
from PIL import Image, ImageEnhance, ImageFilter



# 示例1: 增强亮度
enhancer = ImageEnhance.Brightness(resized_image)
enhanced_image = enhancer.enhance(2.0)  # 将亮度增强2倍
enhanced_image.show()

# 示例2: 增强对比度
enhancer = ImageEnhance.Contrast(resized_image)
enhanced_image = enhancer.enhance(2.0)  # 将对比度增强2倍
enhanced_image.show()

# 示例3: 应用滤镜
filtered_image = resized_image.filter(ImageFilter.SHARPEN)
filtered_image.show()

```
